```python
import matplotlib.pyplot as plt
from PIL import Image
import os
import pandas as pd

image_dir = "./data/raw/aptos2019-blindness-detection/train_images"
csv_path = "./data/raw/aptos2019-blindness-detection/train.csv"

labels_df = pd.read_csv(csv_path).set_index("id_code")

image_files = [f for f in os.listdir(image_dir) if f.endswith(".png")][:9]

fig, axes = plt.subplots(3, 3, figsize=(12, 12))
for ax, fname in zip(axes.flatten(), image_files):
    img = Image.open(os.path.join(image_dir, fname))
    id_code = os.path.splitext(fname)[0]
    label = labels_df.loc[id_code, "diagnosis"]
    ax.imshow(img)
    ax.set_title(f"id: {id_code[:15]} \n label: {label}", fontsize=8)
    ax.axis("off")

plt.tight_layout()
plt.show()
```
